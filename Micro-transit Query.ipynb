```
TripRequests = pandas.read_csv("TRq.csv")
```

```
TripRequests["PickUp_Longitude"] = numpy.array([0 for address in TripRequests["pickup_address"]])
TripRequests["PickUp_Latitude"] = numpy.array([0 for address in TripRequests["pickup_address"]])
TripRequests["DropOff_Longitude"] = numpy.array([0 for address in TripRequests["pickup_address"]])
TripRequests["DropOff_Latitude"] = numpy.array([0 for address in TripRequests["pickup_address"]])
from geopy.geocoders import Nominatim
from random import randint
user_agent = 'user_me_{}'.format(randint(10000,99999))
NmntOj = Nominatim(user_agent=user_agent) #Nominatim Object
```

```
#Get the longitude and latitude for the pick up addres. 
for RNb in range(len(TripRequests)):
    PuLct = NmntOj.geocode(TripRequests.iloc[RNb, 4]) #Pickup location
    if PuLct:
        TripRequests.iloc[RNb, 18] = PuLct.longitude
        TripRequests.iloc[RNb, 19] = PuLct.latitude
    
```

```
#Get the longitude and latitude for the drop off address.
for RNb in range(0, 1951):
    DoLct = NmntOj.geocode(TripRequests.iloc[RNb, 6]) #Dropoff location
    if DoLct:
        TripRequests.iloc[RNb, 20] = DoLct.longitude
        TripRequests.iloc[RNb, 21] = DoLct.latitude
for RNb in range(1952, 4000):
    DoLct = NmntOj.geocode(TripRequests.iloc[RNb, 6]) #Dropoff location
    if DoLct:
        TripRequests.iloc[RNb, 20] = DoLct.longitude
        TripRequests.iloc[RNb, 21] = DoLct.latitude
    
```
